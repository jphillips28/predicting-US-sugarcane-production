### Data Insight Question 2:
How has the weather associated with these sugarcane farming regions changed over the last 40 years?

```python
# Import the regional weather data into pandas DataFrames
df_weather_FL = pd.read_csv("../03 Data Wrangling/df_fl_weather.csv", header=0, index_col=0, \
                            parse_dates=True)
df_weather_FL["State"] = ["FL" for i in df_weather_FL.index]
df_weather_FL["Year"] = df_weather_FL.index.year
df_weather_FL["Month"] = df_weather_FL.index.month
df_weather_FL = df_weather_FL.reset_index()
df_weather_HI = pd.read_csv("../03 Data Wrangling/df_hi_weather.csv", header=0, index_col=0, \
                            parse_dates=True)
df_weather_HI["State"] = ["HI" for i in df_weather_HI.index]
df_weather_HI["Year"] = df_weather_HI.index.year
df_weather_HI["Month"] = df_weather_HI.index.month
df_weather_HI = df_weather_HI.reset_index()
df_weather_LA = pd.read_csv("../03 Data Wrangling/df_la_weather.csv", header=0, index_col=0, \
                            parse_dates=True)
df_weather_LA["State"] = ["LA" for i in df_weather_LA.index]
df_weather_LA["Year"] = df_weather_LA.index.year
df_weather_LA["Month"] = df_weather_LA.index.month
df_weather_LA = df_weather_LA.reset_index()
df_weather_TX = pd.read_csv("../03 Data Wrangling/df_tx_weather.csv", header=0, index_col=0, \
                            parse_dates=True)
df_weather_TX["State"] = ["TX" for i in df_weather_TX.index]
df_weather_TX["Year"] = df_weather_TX.index.year
df_weather_TX["Month"] = df_weather_TX.index.month
df_weather_TX = df_weather_TX.reset_index()

# Combine the daily regional weather DataFrames into one annual regional DataFrame
df_weather = pd.concat([df_weather_FL, df_weather_HI, df_weather_LA, df_weather_TX], \
                       ignore_index=True).set_index("DATE").sort_index()
df_weather_max = df_weather.pivot_table(index="Year", columns="State", values=["TMAX", "WDMV"], aggfunc="max")
df_weather_min = df_weather.pivot_table(index="Year", columns="State", values=["TMIN"], aggfunc="min")
df_weather_sum = df_weather.pivot_table(index="Year", columns="State", values=["PRCP", "SNOW"], aggfunc="sum")
df_weather = pd.concat([df_weather_max, df_weather_min, df_weather_sum], axis=1)
df_weather.to_csv("df_weather.csv")
df_weather.tail()
```
